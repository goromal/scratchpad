# Rotation Matrix

## Construction Techniques

### From Frame Axes $A$ and $B$

```python
import numpy as np
from geometry import SO2
from mesh_plotter.meshes import Axes2DMesh
from mesh_plotter.animator_2d import Animator2D
```

```python
# F = Active

x_A_B = np.array([[-1.0/np.sqrt(2.0), 1.0/np.sqrt(2.0)]]).T
y_A_B = np.array([[-1.0/np.sqrt(2.0), -1.0/np.sqrt(2.0)]]).T

q_A_B = SO2.fromR(np.hstack((x_A_B, y_A_B)))
q_A_A = SO2.identity()

times = [0, 5]
animator = Animator2D(xlim=(-1.5,1.5), ylim=(-1.5,1.5))
animator.addMeshSequence(Axes2DMesh(scale=1.0), [q_A_A, q_A_B], times)
animator.addMeshSequence(Axes2DMesh(scale=1.0), [q_A_A, q_A_A], times)
animator.animate(dt = 0.1, notebook_mode=True)

# Your perspective is glued to the static axes
```

```python
# F = Passive

x_A_B = np.array([[-1.0/np.sqrt(2.0), 1.0/np.sqrt(2.0)]]).T
y_A_B = np.array([[-1.0/np.sqrt(2.0), -1.0/np.sqrt(2.0)]]).T

q_A_B = SO2.fromR(np.vstack((x_A_B.T, y_A_B.T)))
q_A_A = SO2.identity()

times = [0, 5]
animator = Animator2D(xlim=(-1.5,1.5), ylim=(-1.5,1.5))
animator.addMeshSequence(Axes2DMesh(scale=1.0), [q_A_A, q_A_B], times)
animator.addMeshSequence(Axes2DMesh(scale=1.0), [q_A_A, q_A_A], times)
animator.animate(dt = 0.1, notebook_mode=True)

# Your perspective is glued to the moving axes
```
