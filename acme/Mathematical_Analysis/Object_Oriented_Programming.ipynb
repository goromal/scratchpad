## Part 1: Classes

A **class** is a blueprint defining custom objects and their behavior. `__init__()` sets up initial state; **attributes** are stored on `self`.

```python
class Backpack:
    """A simple Backpack with a name and contents list."""
    def __init__(self, name):
        self.name = name
        self.contents = []

    def put(self, item):
        self.contents.append(item)

    def take(self, item):
        self.contents.remove(item)
        

bp = Backpack("Fred")
bp.put("book")
print(bp.name, bp.contents)
```

## Problem 1: Expand the Backpack Class

Add `color`, `max_size=5`, enforce capacity in `put()`, and add `dump()`.

```python
class Backpack:
    """A Backpack with name, color, capacity enforcement, and dump support.

    Attributes:
        name (str): the owner's name.
        color (str): the color of the backpack.
        max_size (int): maximum number of items (default 5).
        contents (list): current items in the backpack.
    """

    def __init__(self, name, color, max_size=5):
        
        """Initialize the backpack with a name, color, and optional max size."""
        self.name = name
        self.color = color
        self.max_size = max_size
        self.contents = []

    def put(self, item):
        """Add item if capacity allows; otherwise print 'No Room!'."""
        if len(self.contents) >= self.max_size:
            print("No Room!")
        else:
            self.contents.append(item)

    def take(self, item):
        """Remove the specified item from contents."""
        self.contents.remove(item)

    def dump(self):
        """Reset contents to an empty list."""
        self.contents = []


def test_backpack():
    testpack = Backpack("Barry", "black")
    assert testpack.name == "Barry"
    assert testpack.color == "black"
    assert testpack.max_size == 5
    for item in ["pencil", "pen", "paper", "computer", "book"]:
        testpack.put(item)
    testpack.put("extra")  # No Room!
    print("Contents:", testpack.contents)
    testpack.dump()
    print("After dump:", testpack.contents)

test_backpack()
```

## Part 2: Inheritance

A **subclass** inherits all methods/attributes from its **superclass**. Use `super()` to delegate to the parent.

```python
class Knapsack(Backpack):
    """A smaller, closeable backpack. Inherits from Backpack."""
    def __init__(self, name, color, max_size=3):
        super().__init__(name, color, max_size)
        self.closed = True

    def put(self, item):
        if self.closed:
            print("I'm closed!")
        else:
            super().put(item)

    def take(self, item):
        if self.closed:
            print("I'm closed!")
            
        else:
            super().take(item)

k = Knapsack("Kyle", "red")
k.put("item")       # closed
k.closed = False
k.put("item")       # succeeds
print(k.contents)
```

## Problem 2: Jetpack Class

Inherit from `Backpack`. Add `fuel`, `fly()`, and override `dump()` to clear both contents and fuel.

```python
class Jetpack(Backpack):
    """A jetpack with fuel that can fly and carry items. Inherits from Backpack.

    Attributes:
        name (str): owner's name.
        color (str): jetpack color.
        max_size (int): item capacity (default 2).
        fuel (int/float): remaining fuel units.
        contents (list): carried items.
    """

    def __init__(self, name, color, max_size=2, fuel=10):
        """Initialize Jetpack with name, color, capacity, and fuel."""
        super().__init__(name, color, max_size)
        self.fuel = fuel

    def fly(self, fuel_amount):
        """Burn fuel_amount to fly. Print 'Not enough fuel!' if insufficient."""
        if fuel_amount > self.fuel:
            print("Not enough fuel!")
        else:
            self.fuel -= fuel_amount

    def dump(self):
        """Empty both the contents list and the fuel tank."""
        self.contents = []
        self.fuel = 0


jp = Jetpack("Janet", "silver")
jp.fly(3)
print(f"Fuel after flying 3: {jp.fuel}")   # 7
jp.fly(100)                                 # Not enough fuel!
jp.put("helmet")
jp.dump()
print(f"After dump — contents: {jp.contents}, fuel: {jp.fuel}")  # [] 0
```

## Problem 3: Add Magic Methods to Backpack

`__eq__`: equal if same name, color, and number of contents. `__str__`: formatted display.

```python
class Backpack:
    """Backpack with equality and string representation support."""

    def __init__(self, name, color, max_size=5):
        self.name = name
        self.color = color
        self.max_size = max_size
        self.contents = []

    def put(self, item):
        if len(self.contents) >= self.max_size:
            print("No Room!")
        else:
            self.contents.append(item)

    def take(self, item):
        self.contents.remove(item)

    def dump(self):
        self.contents = []

    def __eq__(self, other):
        """Equal if same name, color, and number of contents."""
        return (self.name == other.name
                and self.color == other.color
                and len(self.contents) == len(other.contents))

    def __str__(self):
        return (f"Owner:\t\t{self.name}\n"
                f"Color:\t\t{self.color}\n"
                f"Size:\t\t{len(self.contents)}\n"
                f"Max Size:\t{self.max_size}\n"
                f"Contents:\t{self.contents}")


b1 = Backpack("Alice", "blue")
b2 = Backpack("Alice", "blue")
b1.put("book")
b2.put("pen")
print(b1 == b2)   # True
print()
print(b1)
```

## Problem 4: ComplexNumber Class

Implement full complex number arithmetic from scratch.

```python
class ComplexNumber:
    """Complex number a+bj with arithmetic, comparison, and string support."""

    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def conjugate(self):
        """Return the complex conjugate as a new ComplexNumber."""
        return ComplexNumber(self.real, -self.imag)

    def __str__(self):
        if self.imag >= 0:
            return f"({self.real}+{self.imag}j)"
        return f"({self.real}{self.imag}j)"

    def __abs__(self):
        return (self.real**2 + self.imag**2) ** 0.5
        

    def __eq__(self, other):
        return self.real == other.real and self.imag == other.imag

    def __add__(self, other):
        return ComplexNumber(self.real + other.real, self.imag + other.imag)

    def __sub__(self, other):
        return ComplexNumber(self.real - other.real, self.imag - other.imag)

    def __mul__(self, other):
        real = self.real * other.real - self.imag * other.imag
        imag = self.real * other.imag + self.imag * other.real
        return ComplexNumber(real, imag)

    def __truediv__(self, other):
        denom = other.real**2 + other.imag**2
        real = (self.real * other.real + self.imag * other.imag) / denom
        imag = (self.imag * other.real - self.real * other.imag) / denom
        return ComplexNumber(real, imag)


def test_ComplexNumber(a, b):
    py_cnum = complex(a, b)
    my_cnum = ComplexNumber(a, b)
    assert my_cnum.real == a and my_cnum.imag == b
    assert py_cnum.conjugate().imag == my_cnum.conjugate().imag
    assert str(py_cnum) == str(my_cnum), f"{str(py_cnum)} != {str(my_cnum)}"
    print(f"test_ComplexNumber({a}, {b}) passed")

test_ComplexNumber(2, 3)
test_ComplexNumber(5, -1)
test_ComplexNumber(0, 4)

c1 = ComplexNumber(2, 3)
c2 = ComplexNumber(1, -1)
print(c1 + c2)          # (3+2j)
print(c1 * c2)          # (5+1j)
print(f"|{c1}| = {abs(c1):.4f}")   # sqrt(13)
```
